## 2) Análise Exploratória - Introdução

### Explicações Importantes: 20 Cruzamentos, Erros x Hipóteses, Valores das Flags

---

#### **1. O que são os cruzamentos?**

Os cruzamentos são regras analíticas criadas a partir da combinação de variáveis do Censo Escolar, com o objetivo de identificar possíveis incoerências, contradições, lacunas ou padrões atípicos nos dados autodeclarados pelas escolas.  
Cada cruzamento representa uma hipótese de checagem ou um potencial ponto de fragilidade na informação reportada — e dá origem a uma coluna (“flag”) que classifica a situação daquela escola.

---

#### **2. O que diferencia Erros (ERR) de Hipóteses (HIP)?**

- **ERR:** Sinaliza um possível erro factual, incoerência ou contradição explícita no dado, sem justificativa menos plausível. Exemplo: escola informa laboratório de informática, mas não possui energia elétrica.
- **HIP:** Indica uma hipótese de investigação. É uma situação pouco comum ou improvável, mas que pode ser verdadeira em contextos específicos. Exemplo: escola rural sem transporte escolar.

Essa distinção é importante: flags do tipo ERR indicam problemas mais críticos e diretos, enquanto flags HIP servem como alerta para situações que merecem atenção e análise contextual.

---

#### **3. O que significa cada valor das flags?**

- **COERENTE:** Não há indícios de erro ou contradição. O cruzamento está adequado.
- **INCONSISTENTE:** Há sinais claros de erro, contradição ou hipótese relevante para investigação.
- **INCONCLUSIVO:** Não é possível avaliar a consistência, pois ao menos uma variável necessária para o cruzamento está ausente, nula ou vazia. Esses casos são essenciais para identificar o impacto de *missing data*” no acervo de dados.

➡️ Abaixo números absolutos e percetuais para os Tipos, Flags, e Valores.

```python
flags_cols = [
    "STATUS_ERR_FLAG_LABINFO_SEM_ENERGIA",
    "STATUS_HIP_FLAG_RURAL_SEM_TRANSP",
    "STATUS_ERR_FLAG_PRIVADA_VERBA_PUBLICA",
    "STATUS_ERR_FLAG_INFANTIL_SEM_BANHEIRO",
    "STATUS_HIP_FLAG_BIBLIOTECA_SEM_BIBLIOTECARIO",
    "STATUS_ERR_FLAG_INTERNET_SEM_PC",
    "STATUS_ERR_FLAG_ALIMENTACAO_PRIVADA",
    "STATUS_ERR_FLAG_MEDIO_SEM_PROF",
    "STATUS_HIP_FLAG_ESP_SEM_SALA_ADAPTADA",
    "STATUS_ERR_FLAG_ETAPA_SEM_ALUNO",
    "STATUS_HIP_FLAG_URBANA_SEM_ESGOTO",
    "STATUS_ERR_FLAG_ALIMENTACAO_SEM_AGUA",
    "STATUS_ERR_FLAG_REMOTO_SEM_INTERNET",
    "STATUS_HIP_FLAG_PCD_SEM_ACESSIBILIDADE",
    "STATUS_ERR_FLAG_MULTIMIDIA_SEM_ENERGIA",
    "STATUS_HIP_FLAG_RURAL_COM_INTERNET",
    "STATUS_HIP_FLAG_ESTADUAL_ANOS_INICIAIS",
    "STATUS_HIP_FLAG_FEDERAL_INFANTIL",
    "STATUS_HIP_FLAG_ESGOTO_NAO_COM_AGUA",
    "STATUS_ERR_FLAG_CONFLITO_ACESSIBILIDADE"
]
```

```python
status_report = []

for col in flags_cols:
    total = len(df)
    vc = df[col].value_counts(dropna=False)
    n_coerente = vc.get('COERENTE', 0)
    n_inconsistente = vc.get('INCONSISTENTE', 0)
    n_inconclusivo = vc.get('INCONCLUSIVO', 0)

    pct_coerente = 100 * n_coerente / total
    pct_inconsistente = 100 * n_inconsistente / total
    pct_inconclusivo = 100 * n_inconclusivo / total

    pct_incons_e_inconcl = pct_inconsistente + pct_inconclusivo

    # Tipo extraído do nome
    tipo = "ERR" if "_ERR_" in col else ("HIP" if "_HIP_" in col else "")

    status_report.append({
        "Tipo": tipo,
        "Flag": col,
        "# Coerente": n_coerente,
        "# Inconsistente": n_inconsistente,
        "# Inconclusivo": n_inconclusivo,
        "# Total": total,
        "% Coerente": f"{pct_coerente:.0f}%",
        "% Inconsistente": f"{pct_inconsistente:.0f}%",
        "% Inconclusivo": f"{pct_inconclusivo:.0f}%",
        "% Inconsistente + Inconclusivo": f"{pct_incons_e_inconcl:.0f}%"
    })

status_df = pd.DataFrame(status_report)

# Ordenação conforme solicitado
status_df = status_df.sort_values(
    by=["Tipo", "% Inconsistente + Inconclusivo"],
    ascending=[True, False],
    key=lambda col: col if col.name != "% Inconsistente + Inconclusivo"
        else col.str.replace("%","").astype(float)
)
status_df = status_df[
    ["Tipo", "Flag", "# Coerente", "# Inconsistente", "# Inconclusivo", "# Total",
     "% Coerente", "% Inconsistente", "% Inconclusivo", "% Inconsistente + Inconclusivo"]
]
status_df.reset_index(drop=True, inplace=True)

# Função para destacar linhas ERR e HIP
def highlight_tipo(row):
    if row['Tipo'] == 'ERR':
        return ['background-color: #fee3e0'] * len(row)
    elif row['Tipo'] == 'HIP':
        return ['background-color: #e1f5fe'] * len(row)
    else:
        return [''] * len(row)

# Função para negrito apenas na última coluna
def bold_last_col(val):
    return 'font-weight: bold' if isinstance(val, str) and val.endswith('%') else ''

# Aplica ambos: cor na linha, negrito na coluna de soma
styled_df = (
    status_df
    .style
    .apply(highlight_tipo, axis=1)
    .map(bold_last_col, subset=['% Inconsistente + Inconclusivo'])  # Usando .map
)

styled_df
```

## 3) Análise Exploratória - Inconclusivos (Missing Data)

Observa-se que o **percentual de “inconclusivo” é bastante semelhante entre as diferentes flags**, geralmente em torno de 34 a 36 mil escolas. Isso revela uma tendência consistente de **não preenchimento dos campos obrigatórios** por uma parcela significativa das escolas, que aparentemente se limitam ao envio dos dados básicos apenas para cumprir formalmente a obrigação burocrática. A hipótese de preenchimento mínimo é reforçada pela ausência de variação relevante nos números de inconclusivos ao cruzar por diferentes variáveis/ flags.

### Verificação de Aleatoriedade (ou não) em *Missing Data*

A análise de cruzamentos no Censo Escolar revela uma proporção relevante de casos “INCONCLUSIVO”, reflexo de dados ausentes (missing data) nas respostas das escolas.  
No entanto, antes de qualquer estratégia de redistribuição, imputação ou exclusão desses registros, é fundamental **investigar se o padrão de ausência é realmente aleatório** (MCAR), condicional (MAR) ou não aleatório (MNAR).

**Essa etapa é essencial para garantir a robustez das conclusões**:  
Se o missing for sistemático (ex: concentrado em certos perfis de escola, regiões ou condições), qualquer ajuste na análise pode distorcer ou mascarar padrões relevantes.

A seguir, exploramos a distribuição dos casos inconclusivos por diferentes variáveis da base (como dependência administrativa, localização, região e porte escolar), buscando indícios de viés ou padrões estruturais de ausência de dados.

> **Essa investigação permite adotar o tratamento estatístico mais adequado para os registros ausentes, de modo transparente e tecnicamente fundamentado.**


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

xtest = 'TP_DEPENDENCIA_TXT'  # tipo de dependência, como Federal, Estadual, etc.

heatmap_data = []
for flag in flags_cols:
    grupo = df.groupby(xtest)[flag].apply(lambda x: (x == 'INCONCLUSIVO').mean() * 100)
    heatmap_data.append(grupo)

heatmap_df = pd.DataFrame(heatmap_data, index=flags_cols).fillna(0)

plt.figure(figsize=(10, len(flags_cols)*0.45+2))
sns.heatmap(
    heatmap_df,
    annot=True,
    fmt=".1f",
    cmap="Reds",
    linewidths=0.3,
    cbar_kws={"label": "% Inconclusivo"}
)
plt.title(f"Percentual de INCONCLUSIVO por {xtest} e Flag")
plt.xlabel(xtest)
plt.ylabel("Flag")
plt.tight_layout()
plt.show()
```

**Resumo:**  
- Escolas federais têm baixíssimo percentual de inconclusivo (<3%).
- Escolas estaduais apresentam patamar intermediário (~12%).
- Municipais e privadas concentram o maior problema (~17–19%).
- O padrão é igual para todas as flags dentro de cada dependência.

### Conclusão: Tratamento dos Inconclusivos

A análise dos heatmaps evidencia que a ausência de dados (“INCONCLUSIVO”) não é aleatória: há viés estrutural, com maior incidência em escolas rurais, municipais e das regiões Norte/Nordeste. Esse padrão inviabiliza qualquer redistribuição proporcional ou imputação simples, pois tais estratégias mascarariam desigualdades importantes do Censo Escolar.

**Diante desse cenário, adotamos a seguinte abordagem:**

- O indicador principal de inconsistência para cada cruzamento será o **ponto médio** entre duas hipóteses extremas:
  1. Descartar todos os inconclusivos da análise;
  2. Considerar todos os inconclusivos como inconsistentes.
- Essa estratégia comunica explicitamente a incerteza analítica e evita conclusões enganosas — nem otimistas, nem pessimistas.
- Flags com mais de 30% de inconclusivo serão excluídas das análises e dashboards principais, para garantir a robustez e a transparência dos resultados.

> Assim, o trabalho entrega indicadores comparáveis, mas honestos sobre seus limites, permitindo que gestores, pesquisadores e interessados tenham real dimensão da confiabilidade dos dados do Censo Escolar.


## 4) Análise Exploratória - Inconsistentes (Recalibradas)

### Síntese dos Indicadores de Inconsistência

Para cada cruzamento, reportamos três cenários:
- **% Subestimado:** Percentual de inconsistência considerando apenas os registros válidos (descartando os inconclusivos).
- **% Superestimado:** Percentual assumindo que todos os inconclusivos são inconsistentes (cenário pessimista).
- **% MED Inconsistente:** Média das duas hipóteses anteriores, adotada como indicador principal por refletir a incerteza do dado. Essa será a métrica utilizada nos dashboards.

```python
# Calcula os percentuais para cada flag
tabela_report = []

for col in flags_cols:
    total = len(df)
    vc = df[col].value_counts(dropna=False)
    n_coerente = vc.get('COERENTE', 0)
    n_inconsistente = vc.get('INCONSISTENTE', 0)
    n_inconclusivo = vc.get('INCONCLUSIVO', 0)
    total_validos = n_coerente + n_inconsistente

    # % Subestimado: só quem respondeu
    pct_sub = (n_inconsistente / total_validos * 100) if total_validos else 0
    # % Superestimado: todos os missing como inconsistentes
    pct_super = ((n_inconsistente + n_inconclusivo) / total * 100) if total else 0
    # % MED Inconsistente: média aritmética
    pct_med = (pct_sub + pct_super) / 2
    # % Coerente sobre o total
    pct_coerente = n_coerente / total * 100

    tipo = "ERR" if "_ERR_" in col else ("HIP" if "_HIP_" in col else "")

    tabela_report.append({
        "Tipo": tipo,
        "Flag": col,
        "% Coerente": f"{pct_coerente:.0f}%",
        "% Subestimado": f"{pct_sub:.0f}%",
        "% Superestimado": f"{pct_super:.0f}%",
        "% MED Inconsistente": f"{pct_med:.0f}%"
    })

import pandas as pd
df_tabela = pd.DataFrame(tabela_report)

# Ordenação: Tipo (ASC), depois % MED Inconsistente (DESC)
df_tabela = df_tabela.sort_values(
    by=["Tipo", "% MED Inconsistente"],
    ascending=[True, False],
    key=lambda col: col if col.name != "% MED Inconsistente"
        else col.str.replace('%','').astype(float)
).reset_index(drop=True)

# Função para colorir linhas por tipo
def highlight_tipo(row):
    if row['Tipo'] == 'ERR':
        return ['background-color: #fee3e0'] * len(row)
    elif row['Tipo'] == 'HIP':
        return ['background-color: #e1f5fe'] * len(row)
    else:
        return [''] * len(row)

# Exibição estilizada (cor ERR/HIP e negrito só no MED)
styled_table = (
    df_tabela
    .style
    .apply(highlight_tipo, axis=1)
    .map(lambda v: 'font-weight: bold', subset=['% MED Inconsistente'])
)

styled_table
```

## Encerramento: Seleção dos Cruzamentos Prioritários

A partir do indicador sintético **% MED Inconsistente**, definimos critérios objetivos para selecionar os cruzamentos de maior relevância para análise aprofundada:

- **Hipóteses (HIP):** Flags com % MED Inconsistente superior a 30% foram consideradas prioritárias, pois sinalizam situações atípicas com alta magnitude, exigindo investigação por variáveis de contexto.
- **Erros factuais (ERR):** Flags com % MED Inconsistente acima de 10% foram selecionadas, pois representam incoerências ou contradições relevantes no conjunto de dados.

As flags priorizadas compõem a camada `refined_dashboard` e serão cruzadas com as principais variáveis de perfil escolar (região, UF, localização, dependência, porte etc.) nas análises subsequentes em Power BI, permitindo identificar padrões críticos e subsidiar diagnósticos robustos.

**Esse critério de seleção garante foco nos principais desafios da qualidade dos dados autodeclarados, respeitando tanto a magnitude do problema quanto a natureza da inconsistência.**
